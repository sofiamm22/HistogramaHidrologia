# Histograma - Convocatorias Públicas SICEP (2020-2024)

### obtenido de: https://sicep.xm.com.co/public-announcements

##### Antes de correr el código asegurese de descargar el archivo ConvocatoriasPublicas.csv

```python
#Importar librerias necesarias
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Cargar los datos contenidos en el archivo .csv
CP = pd.read_csv('ConvocatoriasPublicas.csv', sep=';', index_col=0, encoding='latin1')
CP.head()
```

### Idea 1 - función .bar

```python
# Contar la frecuencia de cada estado
frecuencia_estados = CP['Estado'].value_counts()

# Crear el histograma (gráfico de barras)
plt.bar(frecuencia_estados.index, frecuencia_estados.values, color='skyblue')

# Agregar etiquetas y título
plt.xlabel('Estado')
plt.ylabel('Frecuencia')
plt.title('Histograma de Estados')

# Rotar etiquetas del eje x para una mejor visualización
plt.xticks(rotation=45)

# Mostrar el histograma
plt.show()
```

### Idea 2 - función .hist

```python
plt.figure(figsize=(10, 6))  # Tamaño del gráfico
CP['Estado'].hist(color='lightseagreen', edgecolor='black', alpha=0.7)  # Crear el histograma

# Agregar etiquetas y título
plt.xlabel('Estado')
plt.ylabel('Frecuencia')
plt.title('Convocatorias Públicas SICEP (2020-2024)')

# Mostrar el histograma
plt.show()
```
